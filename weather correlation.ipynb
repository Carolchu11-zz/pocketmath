## Plot relation between rainy and order for different continent

```python
raw_data = {'continent': ['Europe', 'America', 'China', 'Australia', 'MidEast', 'EastAsia'],
        'order_rain': [len(Europe[(Europe['precip_light']==1) | (Europe['precip_heavy']==1)]), 
                       len(America[(America['precip_light']==1) | (America['precip_heavy']==1)]), 
                       len(China[(China['precip_light']==1) | (China['precip_heavy']==1)]),
                       len(Australia[(Australia['precip_light']==1) | (Australia['precip_heavy']==1)]),
                       len(MidEast[(MidEast['precip_light']==1) | (MidEast['precip_heavy']==1)]),
                       len(EastAsia[(EastAsia['precip_light']==1) | (EastAsia['precip_heavy']==1)])],
        'order_norain': [len(Europe[(Europe['precip_light']==0) & (Europe['precip_heavy']==0)]), 
                         len(America[(America['precip_light']==0) & (America['precip_heavy']==0)]),
                         len(China[(China['precip_light']==0) & (China['precip_heavy']==0)]),
                         len(Australia[(Australia['precip_light']==0) & (Australia['precip_heavy']==0)]),
                         len(MidEast[(MidEast['precip_light']==0) & (MidEast['precip_heavy']==0)]),
                         len(EastAsia[(EastAsia['precip_light']==0) & (EastAsia['precip_heavy']==0)])],
        'order_total': [len(Europe), len(America), len(China), len(Australia), len(MidEast), len(EastAsia)]}
df_rain = pd.DataFrame(raw_data, columns = ['continent', 'order_rain', 'order_norain', 'order_total'])
df_rain
```

```python
# Create the general blog and the "subplots" i.e. the bars
f, ax1 = plt.subplots(1, figsize=(10,5))

# Set the bar width
bar_width = 0.75

# positions of the left bar-boundaries
bar_l = [i+1 for i in range(len(df_rain['order_norain']))]

# positions of the x-axis ticks (center of the bars as bar labels)
tick_pos = [i+(bar_width/2) for i in bar_l]

# Create a bar plot, in position bar_1
ax1.bar(bar_l,
        # using the pre_score data
        df_rain['order_norain'],
        # set the width
        width=bar_width,
        # with the label pre score
        label='Order NoRain',
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#F1911E')

# Create a bar plot, in position bar_1
ax1.bar(bar_l,
        # using the mid_score data
        df_rain['order_rain'],
        # set the width
        width=bar_width,
        # with pre_score on the bottom
        bottom=df_rain['order_norain'],
        # with the label mid score
        label='Order Rain',
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#F4561D')


# set the x ticks with names
plt.xticks(tick_pos, df_rain['continent'])

# Set the label and legends
ax1.set_ylabel("Total Order")
ax1.set_xlabel("Continent")
plt.legend(loc='upper left')

# Set a buffer around the edge
plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
```

## Plot relation between visibility and order for different continent

```python
raw_data = {'continent': ['Europe', 'America', 'China', 'Australia', 'MidEast', 'EastAsia'],
        'order_visibility_low': [len(Europe[Europe['visibility_low']==1]), 
                       len(America[America['visibility_low']==1]), 
                       len(China[China['visibility_low']==1]),
                       len(Australia[Australia['visibility_low']==1]),
                       len(MidEast[MidEast['visibility_low']==1]),
                       len(EastAsia[EastAsia['visibility_low']==1])],
        'order_visibility_high': [len(Europe[Europe['visibility_low']==-1]), 
                         len(America[America['visibility_low']==-1]),
                         len(China[China['visibility_low']==-1]),
                         len(Australia[Australia['visibility_low']==-1]),
                         len(MidEast[MidEast['visibility_low']==-1]),
                         len(EastAsia[EastAsia['visibility_low']==-1])],
        'order_total': [len(Europe), len(America), len(China), len(Australia), len(MidEast), len(EastAsia)]}
df_vis = pd.DataFrame(raw_data, columns = ['continent', 'order_visibility_low', 'order_visibility_high', 'order_total'])
df_vis
```

```python
# Create the general blog and the "subplots" i.e. the bars
f, ax1 = plt.subplots(1, figsize=(10,5))

# Set the bar width
bar_width = 0.75

# positions of the left bar-boundaries
bar_l = [i+1 for i in range(len(df_vis['order_visibility_high']))]

# positions of the x-axis ticks (center of the bars as bar labels)
tick_pos = [i+(bar_width/2) for i in bar_l]

# Create a bar plot, in position bar_1
ax1.bar(bar_l,
        # using the pre_score data
        df_vis['order_visibility_high'],
        # set the width
        width=bar_width,
        # with the label pre score
        label='Order Visibility High',
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='green')

# Create a bar plot, in position bar_1
ax1.bar(bar_l,
        # using the mid_score data
        df_vis['order_visibility_low'],
        # set the width
        width=bar_width,
        # with pre_score on the bottom
        bottom=df_vis['order_visibility_high'],
        # with the label mid score
        label='Order Visibility Low',
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='blue')


# set the x ticks with names
plt.xticks(tick_pos, df_vis['continent'])

# Set the label and legends
ax1.set_ylabel("Total Order")
ax1.set_xlabel("Continent")
plt.legend(loc='upper left')

# Set a buffer around the edge
plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
```

## Plot relation between temperature and order for different continent

```python
def temp_dis(dataframe):
    return [len(dataframe[dataframe['temp']<0]),
            len(dataframe[(dataframe['temp'] >= 0) & (dataframe['temp'] < 10)]),
            len(dataframe[(dataframe['temp'] >= 10) & (dataframe['temp'] < 20)]),
            len(dataframe[(dataframe['temp'] >= 20) & (dataframe['temp'] < 30)]),
            len(dataframe[(dataframe['temp'] >= 30) & (dataframe['temp'] < 40)]),
            len(dataframe[dataframe['temp']>40])]
```

```python
fg = plt.figure(figsize=(10,8))
adj = plt.subplots_adjust(hspace=0.4,wspace=0.4)

sp = plt.subplot(2,3,1)
keys = [0,1,2,3,4,5]
values = temp_dis(Europe)
myDictionary = dict(zip(keys, values))
plt.bar(myDictionary.keys(), myDictionary.values(), color='g')
my_xticks = ['<0', '0-10', '10-20', '20-30', '30-40', '>40']
plt.xticks(keys, my_xticks)
plt.xlabel("Temperature")
plt.ylabel("Order Number")
plt.title('Europe')

sp = plt.subplot(2,3,2)
keys = [0,1,2,3,4,5]
values = temp_dis(America)
myDictionary = dict(zip(keys, values))
plt.bar(myDictionary.keys(), myDictionary.values(), color='g')
my_xticks = ['<0', '0-10', '10-20', '20-30', '30-40', '>40']
plt.xticks(keys, my_xticks)
plt.xlabel("Temperature")
plt.ylabel("Order Number")
plt.title('America')

sp = plt.subplot(2,3,3)
keys = [0,1,2,3,4,5]
values = temp_dis(China)
myDictionary = dict(zip(keys, values))
plt.bar(myDictionary.keys(), myDictionary.values(), color='g')
my_xticks = ['<0', '0-10', '10-20', '20-30', '30-40', '>40']
plt.xticks(keys, my_xticks)
plt.xlabel("Temperature")
plt.ylabel("Order Number")
plt.title('China')

sp = plt.subplot(2,3,4)
keys = [0,1,2,3,4,5]
values = temp_dis(Australia)
myDictionary = dict(zip(keys, values))
plt.bar(myDictionary.keys(), myDictionary.values(), color='g')
my_xticks = ['<0', '0-10', '10-20', '20-30', '30-40', '>40']
plt.xticks(keys, my_xticks)
plt.xlabel("Temperature")
plt.ylabel("Order Number")
plt.title('Australia')

sp = plt.subplot(2,3,5)
keys = [0,1,2,3,4,5]
values = temp_dis(MidEast)
myDictionary = dict(zip(keys, values))
plt.bar(myDictionary.keys(), myDictionary.values(), color='g')
my_xticks = ['<0', '0-10', '10-20', '20-30', '30-40', '>40']
plt.xticks(keys, my_xticks)
plt.xlabel("Temperature")
plt.ylabel("Order Number")
plt.title('MidEast')

sp = plt.subplot(2,3,6)
keys = [0,1,2,3,4,5]
values = temp_dis(EastAsia)
myDictionary = dict(zip(keys, values))
plt.bar(myDictionary.keys(), myDictionary.values(), color='g')
my_xticks = ['<0', '0-10', '10-20', '20-30', '30-40', '>40']
plt.xticks(keys, my_xticks)
plt.xlabel("Temperature")
plt.ylabel("Order Number")
plt.title('EastAsia')
```
